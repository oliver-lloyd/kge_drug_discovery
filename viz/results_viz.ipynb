```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
sns.set_style('darkgrid')
```

```python
# Read results
results = pd.read_csv('../experiment/20240417-193810-simple_biokg.csv')
results.model = 'SimplE'
results.sort_values('metric', ascending=False)
```

```python
sns.boxplot(results, x='model', y='metric')
plt.ylim(0, 0.3)
plt.xticks()
plt.ylabel('Hits@10', size=16)
plt.xlabel('')
plt.xlim(-2, 2)
plt.savefig('figures/hits_at_10.png')
```
